Read (with naked eyes, not computer) the dictionary and find out the definitions of the following 12 columns and copy the text next to the column name
- UNITID: Unit ID for institution        
- INSTNM: Institution name
- CITY: City                    
- ZIP: ZIP code              
- ADM_RATE: Admission rate          
- UGDS: Enrollment of undergraduate certificate/degree-seeking students
- TUITIONFEE_IN: In-state tuition and fees         
- MN_EARN_WNE_P10: Mean earnings of students working and not enrolled 10 years after entry

```python
# reimport the data with above 12 columns (ignore all other variables)
# (Write code here)

df_new = df[["UNITID", "INSTNM", "CITY", "ZIP", "STABBR", "ADM_RATE", "UGDS", "TUITIONFEE_IN", "MN_EARN_WNE_P10"]].copy()
df_new.head(5)
```

```python
# Find out the 10 colleges that are hardest to get in (top 10)
# (Write code here)

df_top = df_new.sort_values(by=["ADM_RATE"], ascending=True).head(10)
df_top
```

```python
# 10. Find out the average in-state tuition of top 10
# (Write code here)

df_top["TUITIONFEE_IN"].mean()
```

```python
# Find out the average earnings of graduates from the top 10
# (Write code here)

df_top["MN_EARN_WNE_P10"].mean()
```

```python
# Find out the 10 colleges that are easiest to get in (bottom 10)
# (Write code here)

df_bottom = df_new.sort_values(by=["ADM_RATE"], ascending=False).head(10)
df_bottom
```

```python
# Find out the average in-state tuition of the bottom 10
# (Write code here)

df_bottom["TUITIONFEE_IN"].mean()
```

```python
# Find out the average earnings of graduates from the bottom 10
# (Write code here)

df_bottom["MN_EARN_WNE_P10"].mean()
```
